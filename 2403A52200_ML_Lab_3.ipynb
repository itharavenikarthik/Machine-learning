```python
x_min=x.min()
x_max=x.max()
x_norm=(x-x_min)/(x_max-x_min)
x_norm
```

```python
y_min=y.min()
y_max=y.max()
y_norm=(y-y_min)/(y_max-y_min)
y_norm
```

```python
m=np.random.randn()
c=np.random.randn()
print("Randomly initiailzed slope:",m)
print("Randomly initiailzed intercept:",c)
```

```python
def predict(x,m,c):
  y_pred=m*x+c
  return y_pred
predict(x,m,c)
```

```python
y_pred=predict(x_norm,m,c)
```

```python
print("First 5 predict values:")
print(y_pred[:5])
```

```python
def compute_cost(y, y_pred):
  n = len(y)
  cost = (1 / (2 * n)) * np.sum((y_pred - y) ** 2)
  return cost
```

```python
m=0.0
c=0.0
y_pred=predict(x_norm,m,c)
cost=compute_cost(y,y_pred)
print("cost for m =",m,"and c =",c,"is:",cost)
```

```python
learning_rate = 0.01
epochs = 1000

n = len(y_norm)
cost_history = []

for epoch in range(epochs):
  y_predict=predict(x_norm,m,c)
  dm=(1/n)*np.sum(x_norm * (y_predict-y_norm))
  dc=(1/n)*np.sum(y_predict-y_norm)

  m=m-learning_rate*dm
  c=c-learning_rate*dc

  cost = compute_cost(y_norm, y_predict)
  cost_history.append(cost)

  if epoch % 100 == 0:
    print(f"Epoch {epoch}, Cost: {cost}")
```

```python
final_slope=m
final_intercept=c
print("Final slope:",final_slope)
print("Final intercept:",final_intercept)
```

```python
def predict_salary(years_experience,m,c,x_min,x_max):
  x=np.array(years_experience).reshape(-1,1)
  x_norm=(x-x_min)/(x_max-x_min)
  salary_pred=m*x_norm+c
  return salary_pred
```

```python
years=5
predicted_salary_norm=predict_salary(years,final_slope,final_intercept,x_min,x_max)
predicted_salary_actual = predicted_salary_norm * (y_max - y_min) + y_min
print(f"predicted salary for {years} years of experiene:")
print(predicted_salary_actual[0][0])
```
