```python
exam_Score_array=df['Exam_Score'].to_numpy()
exam_Score_array
```

```python
exam_Score_array=df['Exam_Score'].to_numpy().reshape(-1,1)
exam_Score_array
```

```python
import numpy as np
mean_Score=np.mean(exam_Score_array)
median_Score=np.median(exam_Score_array)
std_Score=np.std(exam_Score_array)
min_Score=np.min(exam_Score_array)
max_Score=np.max(exam_Score_array)
print(mean_Score)
print(median_Score)
print(std_Score)
print(min_Score)
print(max_Score)
```

```python
import matplotlib.pyplot as plt
plt.xlabel('Exam Score')
plt.ylabel('Number of Students')
plt.hist(exam_Score_array, bins=10, edgecolor='black')
plt.title('Histogram of Exam Scores')
plt.show()
```

```python
scores, counts = np.unique(exam_Score_array, return_counts=True)
plt.bar(scores, counts)
plt.xlabel('Exam Score')
plt.ylabel('Frequency')
plt.title('Histogram of Exam Scores')
plt.show()
```

```python
plt.pie(counts, labels=scores, autopct='%1.1f%%')
plt.title('Exam Score Distribution')
plt.show()
```
